```python
def read_phone_numbers():

    phonebook = {}
    print("\n📖 Welcome to Your Digital Phonebook! 📖")
    print("Enter names and numbers (Press Enter on name to finish)\n")

    while True:
        name = input("🔹 Enter Name: ").strip()
        if name == "":
            break
        number = input(f"📞 Enter Phone Number of {name}: ").strip()

        phonebook[name] = number
        print(f"✅ {name} added!\n")

    print("📒 Phonebook setup complete!\n")
    return phonebook


def print_phonebook(phonebook):
    if not phonebook:
        print("\n📭 Your phonebook is empty!\n")
        return

    print("\n📒 Your Saved Contacts 📒")

    for name, number in phonebook.items():
        print(f"📌 {name:<15} ➝ {number}")

def lookup_numbers(phonebook):
    while True:
        name = input("\n🔍 Lookup Name (Press Enter to exit): ").strip()
        if name == "":
            break
        found_number = phonebook.get(name, None)
        if found_number:
            print(f"📲 {name}'s Number: {found_number}")
        else:
            print(f"❌ {name} not found in the phonebook.")


def phonebook_menu():
    phonebook = read_phone_numbers()

    while True:
        print("\n📜 Phonebook Menu:")
        print("1️⃣ View All Contacts")
        print("2️⃣ Lookup a Number")
        print("3️⃣ Exit")

        choice = input("🔸 Choose an option: ").strip()

        if choice == "1":
            print_phonebook(phonebook)
        elif choice == "2":
            lookup_numbers(phonebook)
        elif choice == "3":
            print("\n👋 Exiting Phonebook. Have a great day!")
            break
        else:
            print("❗ Invalid choice. Please try again.")

phonebook_menu()
```
