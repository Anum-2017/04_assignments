```python
def read_phone_numbers():

    phonebook = {}
    print("\nğŸ“– Welcome to Your Digital Phonebook! ğŸ“–")
    print("Enter names and numbers (Press Enter on name to finish)\n")

    while True:
        name = input("ğŸ”¹ Enter Name: ").strip()
        if name == "":
            break
        number = input(f"ğŸ“ Enter Phone Number of {name}: ").strip()

        phonebook[name] = number
        print(f"âœ… {name} added!\n")

    print("ğŸ“’ Phonebook setup complete!\n")
    return phonebook


def print_phonebook(phonebook):
    if not phonebook:
        print("\nğŸ“­ Your phonebook is empty!\n")
        return

    print("\nğŸ“’ Your Saved Contacts ğŸ“’")

    for name, number in phonebook.items():
        print(f"ğŸ“Œ {name:<15} â {number}")

def lookup_numbers(phonebook):
    while True:
        name = input("\nğŸ” Lookup Name (Press Enter to exit): ").strip()
        if name == "":
            break
        found_number = phonebook.get(name, None)
        if found_number:
            print(f"ğŸ“² {name}'s Number: {found_number}")
        else:
            print(f"âŒ {name} not found in the phonebook.")


def phonebook_menu():
    phonebook = read_phone_numbers()

    while True:
        print("\nğŸ“œ Phonebook Menu:")
        print("1ï¸âƒ£ View All Contacts")
        print("2ï¸âƒ£ Lookup a Number")
        print("3ï¸âƒ£ Exit")

        choice = input("ğŸ”¸ Choose an option: ").strip()

        if choice == "1":
            print_phonebook(phonebook)
        elif choice == "2":
            lookup_numbers(phonebook)
        elif choice == "3":
            print("\nğŸ‘‹ Exiting Phonebook. Have a great day!")
            break
        else:
            print("â— Invalid choice. Please try again.")

phonebook_menu()
```
